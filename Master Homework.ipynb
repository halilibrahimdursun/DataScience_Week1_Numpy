```python
import time
import numpy as np

time_start = time.time()
counter = 0
n = 0
size = 20
M = np.zeros((size, size), dtype = 'float32') # At first we have the sizexsize matrix with zeroes.
while n < size:
    x = size
    y = 0
    while x > 0:                              # With this loop we overwrite our main matrix with matrices which sizes are decreasing in every loop.
        a = np.random.random((x,x))
        b = np.random.random((x,x))
        c = a-b
        d = np.array([c[0,0]])                # This line takes only the first element of diagonal
        counter += 1
        if np.all((-0.1<d)&(d<0.1)):          # If the component is in the interval
            M[y:size, y:size] = c             # Then the related slice of the main matrix is replaced by the down-sizing matrices which are repeatedly created one by one in the loop.
            n += 1                            # This addition is for the outer loop
            y += 1                            # This is used for to find the correct part of the main matrix.
            x -= 1                            # Downsizes the matrix which is repeatedly created in the loop.
        else:
            continue

time_end = time.time()

print(M)

print(f'\nNumber of trials: {counter}')

print(f'\nDiagonal: {[M[i, i] for i in range(size)]}')

print(f'\nTotal running time: {time_end-time_start} seconds')
```
